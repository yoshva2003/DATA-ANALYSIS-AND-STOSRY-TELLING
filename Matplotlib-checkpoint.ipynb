```python
import matplotlib.pyplot as plt
import numpy as np
x=np.array([1,2,5,7])
y=np.array([3,8,1,10])
plt.plot(x,y)
plt.xlabel("overse")
plt.ylabel("runs")
plt.plot(y,marker='o',ms=20)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
x=np.array([1,2,5,7])
y=np.array([3,8,1,10])
plt.plot(x,y)
plt.title("sports Date")
plt.xlabel("overse")
plt.ylabel("runs")
plt.plot(y,marker='o',ms=20)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
x=np.array([1,2,5,7])
y=np.array([3,8,1,10])
font1={'family':'serif','color':'blue','size':20}
font2={'family':'serif','color':'blue','size':15}
plt.xlabel("overse",fontdict=font2)
plt.ylabel("runs",fontdict=font1)
plt.plot(y,marker='o',ms=20)
plt.show()
```
